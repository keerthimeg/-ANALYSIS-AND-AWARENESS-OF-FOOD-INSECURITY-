```python
import pandas as pd;
import matplotlib as plt;
hungerTweets = pd.read_csv('TweetsZeroHunger.csv')
foodInsecurity = pd.read_csv('FoodInsecurity.csv')
foodWastage = pd.read_csv('FoodWastage.csv')
fooInsecuriyCountry = pd.read_csv('foodInsecurityCountry-wise.csv')
```

```python
foddInsecurityCol = foodInsecurity[['County', 'Estimated number food insecure individuals(rounded)']].sort_values(by=['Estimated number food insecure individuals(rounded)'], ascending=False)
foddInsecurityCol.plot(x="County", y="Estimated number food insecure individuals(rounded)", kind="bar", figsize=(20,10), title='Florida County-wise poverty insecurity number').set_ylabel("Poverty insecurity ")
```

```python
foddInsecurityCol = foodInsecurity[['County', 'Food Insecurity rate']].sort_values(by=['Food Insecurity rate'], ascending=False)
foddInsecurityCol.plot(x="County", y="Food Insecurity rate", kind="bar", figsize=(20,10),  title='Florida County-wise poverty insecurity %').set_ylabel("Poverty insecurity in %")
```

```python
foodWastage = foodWastage.sort_values(by=['Food Wastage MKg/year'], ascending=False)
fooInsecuriyCountryCp = fooInsecuriyCountry.copy()
fooInsecuriyCountryCp["People who are food insecure"] = fooInsecuriyCountry["People who are food insecure"]/1000
foodConcat =pd.concat([foodWastage.set_index('Country'),fooInsecuriyCountryCp.set_index('Country')], axis=1, join='inner')
#foodConcat = pd.concat([foodWastage, fooInsecuriyCountry[["Country", "People who are food insecure"]]])
foodConcat.plot(kind="bar", figsize=(20,10),  title='Comparison of Country-wise Food Wastage and Average food Required per year').set_ylabel("Food Quantity in Mkg/year")
```

```python
s = hungerTweets[["Tweet ID", "Date"]].groupby("Date").count()
s.plot(kind="bar", figsize=(20,10), title='Tweet Counts #zeroHunger per day').set_ylabel("Tweet (#zeroHunger) Counts")
```
